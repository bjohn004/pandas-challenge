```python
# Dependencies and Setup
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# define total number of schools in the combined data set
school_total = school_data_complete["school_name"].nunique()

school_total
```

```python
# define total number of students in the combined data set
count_total = school_data_complete["student_name"].count()

count_total
```

```python
# define the total budget of all the schools in the school data dataframe
total_budget = school_data["budget"].sum()

total_budget
```

```python
# determine the average math score from the combined data set
average_math_score = round(school_data_complete["math_score"].mean(),2)

average_math_score
```

```python
# determine the average reading score from the combined data set
average_reading_score = round(school_data_complete["reading_score"].mean(),2)

average_reading_score
```

```python
# determine the number of people passing math, but also include reading_score date for further analysis
count_passing_math_df = school_data_complete.loc[school_data_complete["math_score"] >= 70, ["student_name","math_score", "reading_score"]]

count_math = count_passing_math_df["math_score"].count()

count_math
```

```python
# determinehe number of people passing reading
count_passing_reading_df = school_data_complete.loc[school_data_complete["reading_score"] >= 70, ["student_name", "reading_score"]]
count_reading = count_passing_reading_df["student_name"].count()
count_reading 
```

```python
# determine the percent of people passing math
percent_passing_math = round((count_math / count_total)*100,2)

percent_passing_math
```

```python
# determine the percent of people passing reading
percent_passing_reading = round((count_reading / count_total)*100,2)

percent_passing_reading
```

```python
# determine the number of people passing math and reading using the number of people passing math dataframe
passing_math_reading_df = count_passing_math_df.loc[count_passing_math_df["reading_score"] >= 70 , ["student_name","math_score", "reading_score"]]

passing_math_reading_count = passing_math_reading_df["student_name"].count()

passing_math_reading_count 
```

```python
# determine percent passing both math and reading 
percent_passing_both = round((passing_math_reading_count  / count_total)*100,2)

percent_passing_both 
```

```python
#create summary table
summary_df = pd.DataFrame({"Total Schools": [school_total],
                        "Total Students": count_total,
                        "Total Budget": total_budget,
                        "Average Math Score": average_math_score, 
                        "Average Reading score": average_reading_score, 
                        # "Total Passing Math": count_math, 
                        # "Total Passing Reading": count_reading, 
                        "% Passing Math": percent_passing_math, 
                        "% Passing Reading": percent_passing_reading, 
                        "% Passing Both": percent_passing_both})
summary_df
# NOTE SEE VERY BOTTOM OF NOTEBOOK FOR FORMATTED TABLES
```

# School Summary

```python
# List School names to have general understanding
school_names = school_data["school_name"].unique()

school_names
```

```python
# Cleanup column names, and reorganize columns
school_sort = school_data_complete.rename(columns = {"school_name": "Name of School", "Student ID": "Number of Students", "reading_score": "Reading Score", "math_score": "Math Score"})

school_sorted = school_sort[ ["Number of Students", "Name of School", "Math Score", "Reading Score"]]

school_sorted.head()
```

```python
# Start a DF from the school_data DF to take out school name, type of school, and school budget
# This DF will be merged with the more 'mathematical' DF at the end to create one table grouped by school
school_types = school_data[["school_name", "type", "budget"]]

school_types = school_types.rename(columns = {"school_name" : "Name of School", "type": "Type of School", "budget": "School Budget"})

school_types = school_types.set_index(["Name of School"])

school_types
```

```python
# create table that calculates number of students
# this DF will also be added on to with the mathematical calcs
# ultimately will be merged with school_types DF at the end
number_students = school_sorted.groupby("Name of School").count()

number_students = number_students[["Number of Students"]]

number_students
```

```python
# calc the per student budget
per_student_budget = school_types["School Budget"] / number_students["Number of Students"]

per_student_budget
```

```python
# add the per student budget to the school_types DF
school_types["Per Student Budget"] = per_student_budget

school_types
```

```python
# determine average math score by school, round to 2 decimal digits for cleaner look
avg_math = round(school_sorted.groupby("Name of School").mean(), 2)

avg_math = avg_math["Math Score"]

avg_math
```

```python
# add avg_math to our number_students DF
number_students["Average Math Score"] = avg_math
number_students
```

```python

avg_reading = round(school_sorted.groupby("Name of School").mean(), 2)

avg_reading = avg_reading["Reading Score"]

avg_reading
```

```python
# add avg_reading to our number_students DF
number_students["Average Reading Score"] = avg_reading

number_students
```

```python
# regorganize school_started DF to show passing math scores
passing_math = school_sorted.loc[ school_sorted["Math Score"] >= 70, :]

passing_math
```

```python
# determine % passing math
percent_passing_math = passing_math.groupby("Name of School").count()

percent_passing_math = round((percent_passing_math["Math Score"] / number_students["Number of Students"])*100, 2)

percent_passing_math
```

```python
# add % passing to DF
number_students["% Passing Math"] = percent_passing_math

number_students
```

```python
# regorganize school_started DF to show passing reading scores
passing_reading = school_sorted.loc[ school_sorted["Reading Score"] >= 70, :]

passing_reading
```

```python
# determine % passing reading
percent_passing_reading = passing_reading.groupby("Name of School").count()

percent_passing_reading = round((percent_passing_reading["Reading Score"] / number_students["Number of Students"])*100, 2)

percent_passing_reading
```

```python
# add % passing to DF
number_students["% Passing Reading"] = percent_passing_reading

number_students
```

```python
# regorganize school_started DF to show passing math & reading scores
passing_both = school_sorted.loc[ (school_sorted["Math Score"] >= 70) & (school_sorted["Reading Score"] >= 70), :]

passing_both
```

```python
# determine % passing both
percent_passing_both = passing_both.groupby("Name of School").count()

percent_passing_both = round((percent_passing_both["Reading Score"] / number_students["Number of Students"])*100, 2)

percent_passing_both
```

```python
# add % passing to DF
number_students["% Overall Passing"] = percent_passing_both

number_students
```

```python
# merge school_types & number_students by the index
school_summary_df = pd.merge(school_types, number_students, left_index = True, right_index = True)

# resort the index to be alphabetical
school_summary_df = school_summary_df.sort_values(["Name of School"])

# reorganize column order
school_summary_df = school_summary_df [ ['Type of School', 'Number of Students','School Budget', 'Per Student Budget',
        'Average Math Score', 'Average Reading Score',
       '% Passing Math','% Passing Reading', '% Overall Passing', ]]

# display final table
school_summary_df
# NOTE SEE VERY BOTTOM OF NOTEBOOK FOR FORMATTED TABLES
```

# Top Performing School (By % Overall Passing)

```python
# go back to original school summary DF and sort by overall passing in descending order 
top_school_summary_df = school_summary_df.sort_values(["% Overall Passing"], ascending = False)

top_school_summary_df  = top_school_summary_df.iloc[0:5,:]

top_school_summary_df 
# NOTE SEE VERY BOTTOM OF NOTEBOOK FOR FORMATTED TABLES
```

# Bottom Performing Schools (By % Overall Passing)

```python
# go back to original school summary DF and sort by overall passing in ascending order 
bottom_school_summary_df = school_summary_df.sort_values(["% Overall Passing"], ascending = True)

bottom_school_summary_df  = bottom_school_summary_df.iloc[0:5,:]

bottom_school_summary_df
# NOTE SEE VERY BOTTOM OF NOTEBOOK FOR FORMATTED TABLES
```

# Scores by Per Student Spending

```python
# set up bins/boundaries
bins = [0, 584.9, 629.9, 644.9, 689]

group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# reset new DF to the original school summary DF to allow for manipulation of columns and organize by bins of student spending ranges
student_spending_df = school_summary_df

student_spending_df = student_spending_df [["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

student_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels = group_labels, include_lowest = True)

student_spending_df = round(student_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(),2)

student_spending_df
# NOTE NO FORMATTED TABLE PROVIDED BEYOND THE BELOW TABLE
```
